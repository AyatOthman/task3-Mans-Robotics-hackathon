```python
import pandas as pd 
```

```python
bank_df = pd.read_csv("Bank_Customer.csv")
```

```python
bank_df.isna().sum()
```

```python
bank_df.fillna(bank_df.mean(),inplace = True)
```

```python
bank_df.duplicated().sum()
```

```python
bank_df["duration"].unique()
```

```python
import numpy as np
bank_df["duration"] = pd.to_numeric(bank_df["duration"],errors = 'coerce')
```

```python
bank_df["duration"]= bank_df["duration"].astype(int)
```

```python
sorted_df = bank_df.sort_values(by = "duration")
```
